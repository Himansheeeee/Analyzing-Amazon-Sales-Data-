# Handling Outliers:

- observe for all the numeric columns:
- we will use the IQR method to handle the outliers:


```python
cols=['Order ID','Units Sold','Unit Price','Unit Cost','Total Revenue',
      'Total Cost','Total Profit']
```

```python
for i in cols:
    q1=df[i].quantile(0.25)
    q3=df[i].quantile(0.75)
    iqr=q3-q1
    range_low=q1-1.5*iqr
    range_high=q3+1.5*iqr
    df=df.loc[(df[i]>range_low) & (df[i]<range_high)]
df.shape
```

# BINNING:
- observe two columns for binning

```python
# Do the binning of REVENUE GROUP:

def revenue_grp(x):
    if (x<20000):
        return "Very Low"
    elif (x>=20000 and x<500000):
        return "Low"
    elif (x>=500000 and  x<1500000):
        return "Medium"
    else:
        return "High"
    
```

```python
df['Total_Revenue_Group']=df['Total Revenue'].apply(revenue_grp)
df.head(5)
```

```python
df[['Total Revenue','Total_Revenue_Group']].head()
```

#### BINNING Of Total profit:



```python
df['Total Profit'].max()
```

```python
def profit_grp(x):
    if (x<20000):
        return "Very Low"
    elif (x>=20000 and x<100000):
        return "Low"
    elif (x>=100000 and  x<500000):
        return "Medium"
    else:
        return "High"
```

```python
df['Total_Profit_Group']=df['Total Profit'].apply(profit_grp)
df.head(5)
```
